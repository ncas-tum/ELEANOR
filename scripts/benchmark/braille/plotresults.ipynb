```python
import jax
import jax.numpy as jnp
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
acc_br = jnp.load("results/brailleAccFeLIFV2_Bruno.npy")
acc_v1 = jnp.load("results/brailleAccFeLIFV1_Bruno.npy")
acc_lif = jnp.load("results/brailleAccLIF.npy")
acc_nobr = jnp.load("results/brailleAccFeLIF_noBruno.npy")

loss_br = jnp.load("results/brailleLossFeLIFV2_Bruno.npy")
loss_v1 = jnp.load("results/brailleLossFeLIFV1_Bruno.npy")
loss_lif = jnp.load("results/brailleLossLIF.npy")
loss_nobr = jnp.load("results/brailleLossFeLIF_noBruno.npy")
```

```python
with mpl.style.context("boilerplot.ieeetran"):
    fig_acc, ax_acc = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_acc.plot(acc_br, c="#1b9e77", label="Bruno v2")
    ax_acc.plot(acc_v1, c="#d95f02", label="Bruno")
    ax_acc.plot(acc_nobr, c="#7570b3", label="Vanilla")
    ax_acc.plot(acc_lif, c="#e7298a", label="LIF")
    ax_acc.set_title("Accuracy")
    ax_acc.set_xlabel("Epochs")
    ax_acc.set_ylabel("Accuracy")
    plt.legend()
    plt.savefig("acc_bruno_vs_vanilla.png")

    fig_loss, ax_loss = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_loss.plot(loss_br, c="#1b9e77", label="Bruno v2")
    ax_loss.plot(loss_v1, c="#d95f02", label="Bruno")
    ax_loss.plot(loss_nobr, c="#7570b3", label="Vanilla")
    ax_loss.plot(loss_lif, c="#e7298a", label="LIF")
    ax_loss.set_title("Loss")
    plt.legend()
    plt.savefig("loss_bruno_vs_vanilla.png")
    # plt.show()
```
