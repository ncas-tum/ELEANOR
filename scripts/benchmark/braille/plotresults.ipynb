```python
import jax
import jax.numpy as jnp
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
acc_br = jnp.load("results/brailleAccFeLIF_Bruno.npy")
acc_nobr = jnp.load("results/brailleAccFeLIF_noBruno.npy")
loss_br = jnp.load("results/brailleLossFeLIF_Bruno.npy")
loss_nobr = jnp.load("results/brailleLossFeLIF_noBruno.npy")
```

```python
with mpl.style.context("boilerplot.ieeetran"):
    fig_acc, ax_acc = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_acc.plot(acc_br, label="Bruno")
    ax_acc.plot(acc_nobr, label="Vanilla")
    ax_acc.set_title("Accuracy")
    plt.legend()

    fig_loss, ax_loss = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_loss.plot(loss_br, label="Bruno")
    ax_loss.plot(loss_nobr, label="Vanilla")
    ax_loss.set_title("Loss")
    plt.legend()

    plt.show()
```
