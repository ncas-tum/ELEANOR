```python
import jax
import jax.numpy as jnp
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
acc_her = jnp.load("results/brailleAccHeracles2Bit.npy")
acc_lif = jnp.load("results/brailleAccLIF2Bit.npy")
loss_her = jnp.load("results/brailleLossHeracles2Bit.npy")
loss_lif = jnp.load("results/brailleLossLIF2Bit.npy")
```

```python
with mpl.style.context("boilerplot.ieeetran"):
    fig_acc, ax_acc = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_acc.plot(acc_her, label="Heracles")
    ax_acc.plot(acc_lif, label="LIF")
    ax_acc.set_title("Accuracy")
    plt.legend()

    fig_loss, ax_loss = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_loss.plot(loss_her, label="Heracles")
    ax_loss.plot(loss_lif, label="LIF")
    ax_loss.set_title("Loss")
    plt.legend()

    plt.show()
```
