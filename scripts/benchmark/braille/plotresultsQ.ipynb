```python
import jax
import jax.numpy as jnp
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
# acc_her = jnp.load("results2/brailleAccHeracles2Bit.npy")
acc_her = jnp.load("results/brailleAccLIF3Bit.npy")
acc_lif = jnp.load("results2/brailleAccLIF3Bit.npy")
# loss_her = jnp.load("results2/brailleLossHeracles2Bit.npy")
loss_her = jnp.load("results/brailleLossLIF3Bit.npy")
loss_lif = jnp.load("results2/brailleLossLIF3Bit.npy")
```

```python
with mpl.style.context("boilerplot.ieeetran"):
    fig_acc, ax_acc = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_acc.plot(acc_her, c="#1b9e77", label="Heracles")
    ax_acc.plot(acc_lif, c="#d95f02", label="LIF")
    ax_acc.set_title("Accuracy")
    plt.legend()
    # plt.savefig('acc_heracles2Bit.png')

    fig_loss, ax_loss = plt.subplots(figsize=[3.45, 2.3], dpi=200)
    ax_loss.plot(loss_her, c="#1b9e77", label="Heracles")
    ax_loss.plot(loss_lif, c="#d95f02", label="LIF")
    ax_loss.set_title("Loss")
    plt.legend()
    # plt.savefig('loss_heracles2Bit.png')
    plt.show()
```
