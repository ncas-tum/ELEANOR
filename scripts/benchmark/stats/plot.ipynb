```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib as mpl

palette = sns.color_palette("Dark2", 3)

models = ["LIF", "RLIF", "FeLIF", "Bruno"]

df_forward = pd.read_csv("results/time-forward.csv")
df_backward = pd.read_csv("results/time-backward.csv")

df_forward["Time (s)"] = df_forward["Time"] * 1e-9
df_backward["Time (s)"] = df_backward["Time"] * 1e-9
# df["Peak Memory Usage (MB)"] = df["Peak Memory Usage"] / (1024**2)
# df["Peak Memory Usage (GB)"] = df["Peak Memory Usage"] / (1024**3)
# df.rename(columns={"A": "a", "B": "c"})
mpl.rcParams.update(mpl.rcParamsDefault)
with mpl.style.context("boilerplot.ieeetran"):
    fig, axs = plt.subplots(
        2, len(models), figsize=(3.45 * len(models), 4.6), dpi=200, sharey=True
    )

    for i, model in enumerate(models):
        # sns.lineplot(
        #     df[df["Model"] == model],
        #     x="Sequence Length",
        #     y="Peak Memory Usage (MB)",
        #     hue="Hidden Size",
        #     style="Hidden Size",
        #     markers=True,
        #     # marker='o',
        #     palette=palette,
        #     ax=axs[0,i],
        # )
        ax = sns.lineplot(
            df_forward[df_forward["Model"] == model],
            x="Sequence Length",
            y="Time (s)",
            hue="Hidden Size",
            # marker='o',
            # err_style="bars",
            palette=palette,
            ax=axs[0, i],
        )
        axs[0, i].grid(True)
        axs[0, i].set_title(f"{model}")

        ax = sns.lineplot(
            df_backward[df_backward["Model"] == model],
            x="Sequence Length",
            y="Time (s)",
            hue="Hidden Size",
            # marker='o',
            # err_style="bars",
            palette=palette,
            ax=axs[1, i],
        )
        axs[1, i].grid(True)
plt.suptitle("Memory usage and inference time")
plt.show()
```
